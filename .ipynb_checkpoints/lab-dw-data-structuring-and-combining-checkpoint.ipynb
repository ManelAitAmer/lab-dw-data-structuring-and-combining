In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd
```

```python
url1='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url2='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url3='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
```

```python
df_1=pd.read_csv(url1)
df_1.head()
```

```python
df_1.columns=df_1.columns.str.lower().str.replace(" ","_")
```

```python
df_1.rename(columns = {'st':'state'}, inplace = True) 
```

```python
df_1.info()
```

```python
df_1.dropna(how="all",inplace=True)
```

```python
len(df_1)
```

```python
df_2=pd.read_csv(url2)
df_2.head()
```

```python
df_2.columns=df_2.columns.str.lower().str.replace(" ","_")
```

```python
df_2.rename(columns = {'st':'state'}, inplace = True) 
```

```python
df_2.info()
```

```python
df_2.shape
```

```python
df_2.dropna(inplace=True)
```

```python
len(df_2)
```

```python
df_3=pd.read_csv(url3)
df_3.head()
```

```python
df_3.columns=df_3.columns.str.lower().str.replace(" ","_")
```

```python
df_3.rename(columns = {'st':'state'}, inplace = True) 
```

```python
df_3.shape
```

```python
df_3.info()
```

```python
combined_df_1and2=pd.concat([df_1, df_2])
combined_df_1and2
```

```python
all_df_combined=pd.concat([combined_df_1and2, df_3])
all_df_combined
```

```python
all_df_combined.info()
```

```python
all_df_combined.dropna(how="all", inplace=True)
```

```python
all_df_combined.info()
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd
```

```python
url='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
```

```python
marketing_customer_analysis=pd.read_csv(url)
marketing_customer_analysis.head(40)
```

```python
marketing_customer_analysis=marketing_customer_analysis.drop(columns=['unnamed:_0'])
```

```python
marketing_customer_analysis.info()
```

```python
marketing_customer_analysis.shape
```

```python
response_mapped={'Yes':'1','No':'0'}
marketing_customer_analysis['response'].replace(response_mapped, inplace= True)
```

```python
marketing_customer_analysis.value_counts('gender')
```

```python
marketing_customer_analysis.duplicated().sum()
```

```python
marketing_customer_analysis=marketing_customer_analysis.drop_duplicates()
```

```python
len(marketing_customer_analysis)
```

```python
total_revenue_sum=marketing_customer_analysis.pivot_table(index='sales_channel',values=['total_claim_amount'],aggfunc='sum')
total_revenue_sum.round(2)
```

```python
customer_lifetime_value_per_gender_and_education_sum=marketing_customer_analysis.pivot_table(index=['gender','education'] ,values=['customer_lifetime_value'],aggfunc='sum')
customer_lifetime_value_per_gender_and_education_sum
```
